```python
import requests
import re
from bs4 import BeautifulSoup


url = 'https://www.python.org/'
reqs = requests.get(url)
soup = BeautifulSoup(reqs.text, 'html.parser')

urls = ""
finalurls = []
for link in soup.find_all('a'):
	urls += ((link.get('href'))) +"\n"

matches = re.findall(r"https:\/\/(.+?\...+?\.?\/)|https:\/\/(..+?\.?\/)|http:\/\/(.+?\...+?\.?\/)|http:\/\/(..+?\.?\/)|https:\/\/(.+?\...+?\..+)|https:\/\/(..+?\..+)|http:\/\/(.+?\...+?\..+)|http:\/\/(..+?\..+)", urls)
x = list(matches)
newList = []
for i in x:
	for j in i:
		if j != '':
			newList.append(j)
print(set(newList))
```

```python
import requests
import re
from bs4 import BeautifulSoup 


# Making a GET request
r = requests.get('https://weather.com/weather/tenday/l/Denver+CO?canonicalCityId=9298101d6fcde701a8dae5f27c875376e208cae14ce31a258c225f63573146be')

# Parsing the HTML
soup = BeautifulSoup(r.content, 'html.parser')

# finding element by id
s = soup.findAll('div', class_="DetailsSummary--DetailsSummary--1DqhO DetailsSummary--fadeOnOpen--KnNyF")
snow = soup.findAll('p', class_= "DailyContent--narrative--3Ti6_")
day = re.findall(r"h3.+?>(.+?)<.h3", str(s))
highTemp = re.findall(r"highTempValue.+?>(.+?)<", str(s))
lowTemp = re.findall(r'lowTempValue--.+?>(.+?)<', str(s))
forecast = re.findall(r'DetailsSummary--extendedData--307Ax"?>(.+?)<', str(s))
snowChance = re.findall(r'(Chance of snow.+?)<', str(snow))
count = 0
for i in forecast:
	if 'Snow' in i:
		forecast[forecast.index(i)] = i + "." + snowChance[count]
		count = count+1
merged_list = [(day[i], highTemp[i], lowTemp[i], forecast[i]) for i in range(0, len(day))]


for i in merged_list:
	print(f'Date: {i[0]} \nHigh Temp: {i[1]} \nLow Temp: {i[2]} \n{i[3]}\n')
```

```python
import requests
import re
from bs4 import BeautifulSoup 


# Making a GET request
r = requests.get('https://www.simplyrecipes.com/bran-muffins-recipe-8363742')

# Parsing the HTML
soup = BeautifulSoup(r.content, 'html.parser')

# finding element by id
s = soup.findAll('div', class_="comp recipe-block mntl-block")
#getting the name of the food in the recipe
food = soup.find('h2', class_="comp recipe-block__header text-block")
#Getting the labels like prep time, cooktime, totla time, servings, and yield. And the data for each one.
label = re.findall(r'class="meta-text__label">(.+?)<', str(s))
data = re.findall(r'class="meta-text__data">(.+?)<', str(s))
#combing the labels and data.
metadata = [(label[i], data[i]) for i in range(0, len(label))]
#getting the ingreidents
ingredients = soup.find('ul', class_="structured-ingredients__list text-passage")
texts = ingredients.find_all('p')
#getting the explination for each step
stepsExplain = soup.find('li', class_="comp mntl-sc-block-group--LI mntl-sc-block mntl-sc-block-startgroup")

#grouping the step header and the explaination
zeug = [x.get_text() for x in soup.find_all("span", class_="mntl-sc-block-subheading__text")]
zeugs = [x.get_text() for x in soup.find_all("p", class_="comp mntl-sc-block mntl-sc-block-html")]
steps = [(zeug[i], (zeugs[12:])[i]) for i in range(0, len(zeug))]
#number for each step
number = 1
#printing out all of the information.
print(food.text)
for i in metadata:
    for j in i:
        print(j)
    print("\n")
for i in steps:
    print(number)
    for j in i:
        print(j)
    print("\n")
    number = number +1
print("Ingredients:" + "\n")
for text in texts:
    print(text.get_text())
```
